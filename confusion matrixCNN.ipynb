```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Function to create synthetic confusion matrix
def synthetic_confusion_matrix(accuracy, precision, recall, total_samples=1000, classes=3):
    # Calculate TP, FP, FN based on metrics
    tp = int(total_samples * accuracy / 100)  # True Positives
    fn = int(tp / recall * (1 - recall))  # False Negatives
    fp = int(tp / precision * (1 - precision))  # False Positives
    tn = total_samples - tp - fn - fp  # True Negatives
    
    # Split values across classes (simplified)
    confusion_matrix = np.array([
        [tp // classes, fp // classes, fn // classes],
        [fp // classes, tn // classes, fn // classes],
        [fn // classes, fp // classes, tp // classes]
    ])
    return confusion_matrix
```

```python
# Experiments data
experiments = [
    {"name": "First Experiment", "accuracy": 5.71, "precision": 0.1905, "recall": 0.2222},
    {"name": "Second Experiment", "accuracy": 56.01, "precision": 0.739854, "recall": 0.567864},
    {"name": "Third Experiment", "accuracy": 73.89, "precision": 0.487546, "recall": 0.557654},
    {"name": "Using Custom Dataset", "accuracy": 96.4554, "precision": 0.974235, "recall": 0.932314},
    {"name": "30 Epochs (Extended Training)", "accuracy": 0.7867, "precision": 0.9876, "recall": 0.5098},
]
```

```python
# Generate and display confusion matrices
for exp in experiments:
    cm = synthetic_confusion_matrix(
        accuracy=exp["accuracy"],
        precision=exp["precision"],
        recall=exp["recall"]
    )
    disp = ConfusionMatrixDisplay(cm, display_labels=["Class 1", "Class 2", "Class 3"])
    disp.plot(cmap='Blues')
    plt.title(f"Confusion Matrix - {exp['name']}")
    plt.show()
```
